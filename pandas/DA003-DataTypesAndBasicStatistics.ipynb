# 3. Data Types & Basic Statistics

## Key Concepts
### 1. Data Type Conversion
- `pd.to_numeric()` - converts a Series to numeric, handling error (e.g., `coerce` for invalid entries)
- `pd.to_datetime()` converts a Series to datetime objects
- `.astype()` - used for simple type changes (e.g., `float` or `int`)
### 2. Why Convert?
- Math operations (like subtraction for `Profit`) require numbers, not strings
- Time-series analysis requires datetime objects
- Visualizations and statistical functions expect proper data types
### 3. Descriptive Statistics
- **Mean** - arithmetic average ( sum / count ). Sensitive to outliers.
- **Meidan** - middle value (50th percentile). Robust to outliers.
- **Mode** - most frequent value. Used for categorical data (or descrete numeric).
- **Standard Deviation** (`std`) - measure of spread around the mean.
- **Quantiles** - cut points dividing the data into intervalues (e.g., 25%, 50%, 75%). The **Interquartile Range (IQR)** = 75% - 25%.

- Open your notebook and load the file you saved

```python
import pandas as pd
import numpy as np
movies = pd.read_csv("movies_cleaned.csv")

# Double-check that no missing values remain
print(movies.isnull().sum().sum()) # Should be 0
print(movies.dtypes)
```

```python
movies.isnull().sum().sum()
```

### Step 1: Convert `Gross` to numeric (in millions)

- Important: The original `Gross` is in **millions** (e.g., "`$2,796.30`" = 2,796,30 million dollars).
- We will convert `Gross` to **raw dollars** (multiply by 1,000,000) so that `Profit` calculations match the original dataset's scale.

```python
# Clean Gross: remove '$', commas, spaces, convert to float, then multiply by 1000000
movies["Gross"] = movies["Gross"].astype(str).str.replace("$", "")
movies["Gross"]
```

```python
movies["Gross"] = movies["Gross"].str.replace(",", "")
movies["Gross"]
```

```python
movies["Gross"] = movies["Gross"].str.strip()
movies["Gross"]
```

```python
movies["Gross"] = pd.to_numeric(movies["Gross"]) * 1_000_000
movies["Gross"]
```

```python
movies[["Title", "Gross", "WeekendGross"]].head(3)
```

- **Why multiply by 1,000,000?** The original `Profit` column had values like 2847000000 (dollars). To recalculate `Profit` correctly, we need `Gross` in raw dollars.

### Step 2: Convert `ReleaseDate` to datetime

```python
movies['ReleaseDate'] = pd.to_datetime(movies['ReleaseDate'])

# Check the data type
print(movies['ReleaseDate'].dtype)
# Should be datetime64[ns]
print(movies['ReleaseDate'].head(3))
```

- Note: We can now extract the exact weekday name or month name easily.

### Step 3: Recalculate `Profit` and `ROI`

- Now, the `Gross` and `Budget` are numeric, let's `Profit` and `ROI` columns

```python
# Profit = Gross - Budget
movies['Profit'] = movies['Gross'] - movies['Budget']

# ROI = Profit / Budget (as a ratio, not percentage)
movies['ROI'] = movies['Profit'] / movies['Budget']

# Check first 5
movies[["Title", "Budget", "Gross", "Profit", "ROI"]].head()
```

### Step 6: Calculate specific statistics manually (for deeper understanding)

```python
gross_mean = movies["Gross"].mean()
gross_median = movies["Gross"].median()
gross_std = movies["Gross"].std()
gross_quantiles = movies["Gross"].quantile([0.25, 0.50, 0.75])

print("Gross Mean:", gross_mean)
print("Gross Median:", gross_median)
print("Gross Std:", gross_std)
print("Gross Quantiles:", gross_quantiles)
```

- **Interpretation**: The IQR = 75% - 25%. This tells us the middle 50% of movies have grosses that differ.

### Step 8: Understand skewness with mean vs median

- Right-skewed: mean > median
- Left-skewed: mean < median
- Symmetric: mean = median

```python
cols = ["Gross", "Budget", "ROI", "Runtime", "Votes"]
for col in cols:
    mean_val = movies[col].mean()
    median_val = movies[col].median()
    skew = "Right-skewed" if mean_val > median_val else "Left-skewed" if mean_val < median_val else "Symmetric"
    print(col + " | " + "Mean: " + str(mean_val) + " | " + str(median_val) + " | " + str(skew))
```

- **Why it matters**: Right-skew means a few high-value outliers pull the mean up. When reporting averages, **median** is often a better summary for skewed data.

### Step 9: Quantile practice - find the 90th percentile

- You might want to know: "What gross is required to be in the top 10%?"

```python
top_10_percent_gross = movies['Gross'].quantile(0.90)
```

```python
print("90th percentile gross:", top_10_percent_gross)
```

```python
# Count how many movies exceed this
(movies['Gross'] > top_10_percent_gross).sum()
```
