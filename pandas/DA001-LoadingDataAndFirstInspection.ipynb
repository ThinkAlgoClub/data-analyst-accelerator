### Step 5: Numerical Summary with `.describe()`

````python
movies.describe()  # descriptive statistics
````

##### **Key Statistics**
- **Count** - Number of non-missing values
- **Mean vs Median (50%)** - If they differ, the distribution is skewed
- **Min/Max** - identify extreme values
- **Std** - standard deviation; larger means more spread
---
**NOTE**: `describe()` automatically excludes *object* columns. To see the summary for categorical columns, we use `.describe(include='object')`


````python
movies.describe(include='object')
````

### Step 6: Check the first and last few rows with `.head()` and `.tail()`

````python
movies.tail(5)   # Last 5 rows
````

- **Why view tail?** Sometimes data is sorted in a spcific order; looking at the end helps confirm that the dataset covers the entire top-100 list correctly.

### Step 9: Quick data exploration - Unique values in categorical columns

````python
movies.Title   # movies.Studio
````

````python
print("Unique MPAA ratings:", movies["MPAA"].unique())
````

````python
print("Unique Languages:", movies['Language'].unique())
````

- **Why this matters?** Knowing the unique values helps to decide how to encode, group, or filter data later.

### Step 10: Memory usage

````python
print("Memory usage (in MB):", movies.memory_usage(deep=True).sum() / 1024**2)
````

(This is tiny, but in real datasets memory is crucial.)

#### Golden rule:
- Always run `.info()` and `.describe()` as soon as you load new data - they give you 80% of what you need to know before doing any analysis.
