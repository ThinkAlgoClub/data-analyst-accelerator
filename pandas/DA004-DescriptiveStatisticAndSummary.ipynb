# Descriptive Statistics & Summary

## Key Concepts
### 1. `.describe()` - The Swiss Army Knife of Summary
- **Numeric columns**: Count, mean, std, min, 25%, 50%, 75%, max.
- **Object (categorical) columns**: Count, unique, top (mode), freq (frequency of top value)
- `include='all'`: Forces pandas to show both numeric and categorical summaries in the table.
### 2. Frequency Tables ( `.value_counts()`)
- Absolute counts: `df['col'].value_counts()`
- Relative frequencies (percentages): `df['col'].value_counts(normalize=True)
- Sort order is descending by count by default.
### 3. Grouped Descriptives
- `df.groupby('category')['numeric'].descriptive()
    - great for comparing distributions across groups (e.g., PG-13 vs R movies)
### 4. Skewness & Kurtosis
- **Skewness** (`df['col'].skew()`):
    - 0 = symmetric (normal-like)
    - > 0 = right-skewed (long right tail)
    - < 0 = left-skewed (long left tail)
- **Kurtosis** (`df['col']kurtosis()`):
    - 3 = heavy tails (more outliers)
    - <3 = light tails (fewer outliers)
    - (Often used with normal distribution benchmark of 0 if using excess kurtosis, but pandas uses Fisher's excess kurtosis where normal=0).
### 5. Cross-tabulation ( `pd.crosstab()`)
- Show the frequency of combinations of two categorial variables (e.g., `Sequel` vs `MPAA`).

### Step 0: Load the final dataset

- Load the fully cleaned file:

```python
import pandas as pd
import numpy as np
movies = pd.read_csv('movies_final.csv')

# Quick sanity check
print("Shape:", movies.shape)
print("\nFirst 2 rows:")
movies.head(2)
```

### Step 1: Classic `.describe()` for numeric columns

```python
numeric_summary = movies.describe()
numeric_summary
```

- **Add a useful customization**: Exclude columns you don't care about (like `Rank` or `Age`. if they don't make sense for stats):

```python
# Describe only a subset of numeric columns
movies[['Gross', 'Budget', 'ROI', 'Runtime']].describe()
```

### Step 2: `.describe()` for categorical (object) columns

```python
categorical_summary = movies.describe(include=['object'])
categorical_summary
```

### Step 3: Use `include='all'` for a combined view

```python
# Get both numeric and categorical stats in one table (with NaNs for incapable metrics)
movies.describe(include='all')
```

- This produces a wide table.
- **Highlight**: For numeric columns, you see mean/std; for categorical, you see unique/top/freq. This is great for a quick "everything" summary.

### Step 4: Frequency tables - absolute and relative

- Let's drill down into categorical columns with `.value_counts()`

```python
# Absolute counts
mpaa_counts = movies['MPAA'].value_counts()
print("MPAA Ratings:\n", mpaa_counts)
```

```python
# Relative percentages
mpaa_pct = movies['MPAA'].value_counts(normalize=True)
print("MPAA Ratings (percentage):\n", mpaa_pct.round(3))
```
